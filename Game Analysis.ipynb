```python
num_pitchers = len(data['Pitcher ID'].unique())  
fig, axes = plt.subplots(nrows=num_pitchers, ncols=1, figsize=(8, 4 * num_pitchers)) 

for i, (pitcher_id, group) in enumerate(data.groupby('Pitcher ID')):
    pitcher_first_name = pitchers.loc[pitchers['Pitcher ID'] == pitcher_id, 'Pitcher First Name'].iloc[0]
    pitcher_last_name = pitchers.loc[pitchers['Pitcher ID'] == pitcher_id, 'Pitcher Last Name'].iloc[0]
    pitcher_name = f"{pitcher_first_name} {pitcher_last_name}"
    ax = axes[i]  # Get the appropriate subplot for the pitcher
    group.plot(x='Pitch Number', y='Pitch TotalSpeed (MPH)', ax=ax, label=f'Pitcher {pitcher_name}')  
    ax.set_xlabel('Pitch #')
    ax.set_ylabel('Pitch TotalSpeed (MPH)')
    ax.set_title(f'Velocity of Fastballs - {pitcher_name}')
    ax.legend()   

plt.tight_layout()
plt.show()
```

```python
# Group the data by pitcher and pitch type
grouped = data.groupby(['Pitcher ID', 'Pitch Type'])

# Get the unique pitchers
pitchers = data['Pitcher ID'].unique()

# Create a figure and axes
fig, axes = plt.subplots(nrows=len(pitchers), ncols=1, figsize=(10, 6 * len(pitchers)))

# Iterate over each pitcher
for i, pitcher_id in enumerate(pitchers):
    # Check if the current pitcher has data for any pitch type
    if pitcher_id in grouped.groups:
        # Get the data for the current pitcher
        pitcher_data = grouped.get_group(pitcher_id)
        
        # Get the unique pitch types for the current pitcher
        pitch_types = pitcher_data['Pitch Type'].unique()
        
        # Get the corresponding subplot for the current pitcher
        ax = axes[i]
        
        # Iterate over each pitch type
        for pitch_type in pitch_types:
            # Get the data for the current pitch type
            pitch_type_data = pitcher_data[pitcher_data['Pitch Type'] == pitch_type]
            
            # Plot the pitch velocity
            ax.plot(pitch_type_data['Pitch Number'], pitch_type_data['Pitch TotalSpeed (MPH)'], label=pitch_type)
        
        # Set the x and y-axis labels for the current subplot
        ax.set_xlabel('Pitch Number')
        ax.set_ylabel('Pitch TotalSpeed (MPH)')
        
        # Set the plot title for the current subplot
        ax.set_title(f'Pitcher {pitcher_id} - Velocity of Different Pitch Types')
        
        # Add a legend for the current subplot
        ax.legend()
    else:
        # Remove the subplot if the pitcher doesn't have any data
        fig.delaxes(axes[i])

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
